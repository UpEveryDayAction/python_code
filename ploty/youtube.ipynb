```python
import plotly.express as px
df = px.data.tips()
fig = px.scatter(df, x="total_bill", y="tip", facet_col="smoker", facet_row="sex")
fig.add_annotation(text="paper", showarrow=False, xref="paper", yref="paper", x=0.2, y=0.8)
fig.add_annotation(text="domain", showarrow=False, xref="x domain", yref="y domain", x=0.5, y=0.8) #割合で示す
fig.add_annotation(text="data", showarrow=False, xref="x", yref="y", x=10, y=6)
fig.show()
```

```python
# import plotly.express as px
import plotly.graph_objects as go
df_stocks = px.data.stocks()
px.line(df_stocks, x='date', y=['GOOG','AMZN'], labels={'x':'Date','y': 'Price'},title="stock's price")
# px.line(df_stocks, x='date', y=['GOOG','AMZN'], labels={'x':'Date','y': 'Price'})
# px.line(df_stocks, x='date', y='GOOG', labels={'x':'Date','y': 'Price'})

fig = go.Figure()
fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.AMZN,mode='lines+markers',name='Amazon'))
```

```python
# ラインの装飾について
# fig.add_trace(go.Scatter(x=df_stocks.date, y=df_stocks.GOOG,mode='lines+markers',name='Google', line=dict(color='firebrick', width=2, dash='dashdot')))

fig.update_layout(title='Stock Price Data 2018 - 2020', xaxis_title='Price',yaxis_title='Date')

fig.update_layout(
    xaxis=dict(
showline=True, showgrid=False,showticklabels=True,
linecolor='rgb(204, 204, 204)',
linewidth=2, ticks='outside',tickfont=dict(
    family='Arial',size=12,color='rgb(82,82,82)',
),
),
yaxis=dict(showgrid=False, zeroline=False, showline=False,
showticklabels=False),
autosize=False,
margin=dict(
    autoexpand=False,l=100,r=20, t=110, b=70),
    showlegend=False, plot_bgcolor='white')
```
